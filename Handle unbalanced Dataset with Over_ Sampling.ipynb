```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv('creditcard.csv')
data1.head()
```

```python
data1.info()
```

```python
data1.describe()
```

```python
# put all columns name into a list

columns = data1.columns.tolist()
columns
```

```python
# create independent and dependent feature

columns = [c for c in columns if c not in ['Class']]
columns
```

```python
## Store the variable we are predicting
target = 'Class'
```

```python
## Define a random state

state = np.random.RandomState(45)

X = data1[columns]
Y = data1[target]

X_outlier = state.uniform(low =0, high =1, size =(X.shape[0], X.shape[1]))
```
